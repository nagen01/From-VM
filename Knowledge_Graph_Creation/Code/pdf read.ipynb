```python
import PyPDF2
from zipfile import ZipFile
import os
```

```python
# Create a ZipFile Object and load sample.zip in it
with ZipFile('/home/nb01/C_Drive/Knowledge_Graph_Creation/Dataset/open-test-data-master-rc-1.0.0-1-data-3-provided-USGS_docs.zip', 'r') as zipObj:
   # Extract all the contents of zip file in different directory
   zipObj.extractall('/home/nb01/C_Drive/Knowledge_Graph_Creation/Dataset/PDFs')
```

```python
data_dir ='/home/nb01/C_Drive/Knowledge_Graph_Creation/Dataset/PDFs/' 
print(os.listdir(data_dir))
pdf_text = [0]
for f in os.listdir(data_dir):
    print(f)
    fl = open(data_dir+f, "rb")
    print(fl)
    pdf_reader = PyPDF2.PdfFileReader(fl)
    for p in range(pdf_reader.numPages):
        #print(p)
        page = pdf_reader.getPage(p)
        #print(page)
        pdf_text.append(page.extractText())
        print(pdf_text)
    fl.close()
```
