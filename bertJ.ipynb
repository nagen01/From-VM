```python
import torch
from transformers import BertModel, BertTokenizer
```

```python
#BERT_MODEL_CLASSES = [BertModel, BertForPreTraining, BertForMaskedLM, BertForNextSentencePrediction,
#                      BertForSequenceClassification, BertForTokenClassification, BertForQuestionAnswering]
#BERT_TOKENIZER     = [BertTokenizer]
#pretrained_weights = 'bert-base-uncased'
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased', output_hidden_states = True, output_attentions = True )
```

```python
input_id = torch.tensor([tokenizer.encode("Let's see all hidden-states and attentions on this text", add_special_tokens=True)])
print(input_id)
with torch.no_grad():
    last_hidden_states = model(input_id)
    print("\n*-3=================")
    print(last_hidden_states[-3])
    print("\n*-2=================")
    print(last_hidden_states[-2])
    print("\n*-2=================")
    print(last_hidden_states[-1])
    print("\n*0=================")
    print(last_hidden_states[0])
    print("\n*1=================")
    print(last_hidden_states[1])
    print("\n*2=================")
    print(last_hidden_states[2])
```
